```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table= soup.find('table', 'table table-striped table-bordered')
```

```python
table
```

```python
world_columns = table.find_all('th')
```

```python
world_column_titles = [column.text for column in world_columns]
```

```python
world_column_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_column_titles)

df
```

```python
table.find_all('td')
```

```python
table.find_all('tr')
```

```python
world_rows = table.find_all('tr')
```

```python

for row in world_rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

df
    
```
